# Outlier Detection (IQR Method)

```python
# Outlier Detection using IQR

# Function to detect outliers using IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers.shape[0], lower_bound, upper_bound

# Checking for outliers in key numerical columns
outlier_summary = {col: detect_outliers(df, col) for col in num_cols}
outlier_summary
```

# Remove outliers

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("theatre_seat_prediction.csv")

# Keep the "city" column as it is
columns_to_clean = ["average_ticket_price", "total_seats", "ticket_sold", "budget", "revenue"]

# Function to remove outliers using IQR
def remove_outliers_iqr(df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1

    # Define bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers
    df_cleaned = df[~((df[columns] < lower_bound) | (df[columns] > upper_bound)).any(axis=1)]
    return df_cleaned

# Apply outlier removal
df_cleaned = remove_outliers_iqr(df, columns_to_clean)

# Save the final cleaned dataset
df_cleaned.to_csv("theatre_seat_prediction_no_outliers.csv", index=False)

print("âœ… Final cleaned dataset saved as 'theatre_seat_prediction_no_outliers.csv'")
```

3. Experimenting more with various encoding techniques to check on correlation to find relvant features w.r.t ticket_sold column

```python
# Reload the dataset
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from collections import defaultdict
df_encoded = pd.read_csv("theatre_seat_prediction_no_outliers.csv")

# Identify categorical columns
# categorical_cols = df_encoded.select_dtypes(include=['object']).columns
categorical_cols = [
    "city", "weather", "showtime", "movie_name", "genre","cast", "production",
    "holiday_event_impact" ]

# Dictionary to store encoding methods used
encoding_methods = {}

# Process each categorical column
for col in categorical_cols:
    unique_values = df_encoded[col].nunique()

    if unique_values <= 5:
        # One-Hot Encoding for low-cardinality categorical variables
        encoded_df = pd.get_dummies(df_encoded[col], prefix=col, drop_first=True)
        df_encoded = df_encoded.drop(columns=[col]).join(encoded_df)
        encoding_methods[col] = "One-Hot Encoding"

    elif unique_values <= 15:
        # Ordinal Encoding for medium-cardinality categorical variables
        ordinal_encoder = OrdinalEncoder()
        df_encoded[col] = ordinal_encoder.fit_transform(df_encoded[[col]])
        encoding_methods[col] = "Ordinal Encoding"

    else:
        # Frequency Encoding for high-cardinality categorical variables
        freq_encoding = df_encoded[col].value_counts().to_dict()
        df_encoded[col] = df_encoded[col].map(freq_encoding)
        encoding_methods[col] = "Frequency Encoding"

# Save the encoded dataset
encoded_file_path = "theatre_seat_prediction_encoded.csv"
df_encoded.to_csv(encoded_file_path, index=False)

#import os
#os.path.exists(encoded_file_path)

```

Q - Is holiday_event_impact column have relationship with ticket_sold ?

Ans - Rejected Null Hypothesis + Found Negative Correlation

Point-Biserial Correlation Output

If positive, ticket sales increase on holidays.

If negative, fewer tickets are sold on holidays.

If close to zero, no clear relationship.

T-test p-value Output

p < 0.05 â†’ Holidays significantly affect ticket sales.

p > 0.05 â†’ No strong statistical difference.

```python
import pandas as pd
import scipy.stats as stats

# Load the dataset
file_path = "theatre_seat_prediction_encoded.csv"
df = pd.read_csv(file_path)

# Convert 'holiday_event_impact' to numeric
df["holiday_event_impact"] = pd.to_numeric(df["holiday_event_impact_Yes"], errors='coerce')

# Check unique values and handle potential issues
unique_values = df["holiday_event_impact"].nunique()
if unique_values <= 1:
    print("Warning: 'holiday_event_impact' has only one unique value. Correlation and t-test cannot be performed.")
    correlation_value, t_test_p_value = None, None  # Explicitly set to None
elif df["holiday_event_impact"].isnull().all():
    print("Warning: 'holiday_event_impact' contains only NaN values after conversion. Correlation and t-test cannot be performed.")
    correlation_value, t_test_p_value = None, None  # Explicitly set to None
else:
    # Calculate Point-Biserial Correlation
    corr_holiday = stats.pointbiserialr(df["holiday_event_impact"], df["ticket_sold"])
    correlation_value = corr_holiday.correlation

    # Perform a t-test
    holiday_sales = df[df["holiday_event_impact"] == 1]["ticket_sold"]
    non_holiday_sales = df[df["holiday_event_impact"] == 0]["ticket_sold"]

    if len(holiday_sales) > 1 and len(non_holiday_sales) > 1:
        t_test_p_value = stats.ttest_ind(holiday_sales, non_holiday_sales, equal_var=False).pvalue
    else:
        print("Warning: Not enough data for t-test in one or both groups.")
        t_test_p_value = None

correlation_value, t_test_p_value  # Print the results
```

# Feature Engineering Options Based on Correlation Analysis

To improve predictions and insights, we can create new features and transform existing ones:

Interaction Features (Multiplication / Ratios)

Revenue per Seat: revenue / total_seats â†’ Measures efficiency of seat usage.

Budget Efficiency: revenue / budget â†’ Determines how well a budget converts to earnings.

Price Sensitivity: average_ticket_price * ticket_sold â†’ Checks impact of pricing on sales.

Categorical Encoding (For Machine Learning Models)

Convert categorical columns (genre, weather, city) into one-hot encoding or ordinal encoding.

Log Transformation (For Skewed Data)
log(budget), log(revenue), and log(ticket_sold) can stabilize extreme values.

Time-Based Features (From Showtime & Weather)

Create "Peak Hours" category (e.g., Morning, Afternoon, Evening).

Convert weather conditions into binary groups (favorable vs. unfavorable).

```python
import numpy as np

# Step 1: Create Feature Engineering Columns

# Interaction Features
df["revenue_per_seat"] = df["revenue"] / df["total_seats"]
df["budget_efficiency"] = df["revenue"] / df["budget"]
df["price_sensitivity"] = df["average_ticket_price"] * df["ticket_sold"]

# Log Transformations (to reduce skewness)
df["log_budget"] = np.log1p(df["budget"])
df["log_revenue"] = np.log1p(df["revenue"])
df["log_ticket_sold"] = np.log1p(df["ticket_sold"])

# Time-based Feature (Peak Hours from Showtime)
df["peak_showtime"] = df["showtime"].apply(lambda x: "Peak" if x in ["Evening", "Night"] else "Off-Peak")

# Weather Binary Categorization
df["favorable_weather"] = df["weather"].apply(lambda x: 1 if x in ["Clear", "Partly Cloudy"] else 0)

# Handling Inf/NaN values caused by division
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Fill NaN values with column median (better than mean for skewed data)
df.fillna(df.median(numeric_only=True), inplace=True)

# Verify if any inf/NaN values remain
df.isna().sum()

# Save updated dataset with new features
updated_file_path = "theatre_seat_prediction_engineered.csv"
df.to_csv(updated_file_path, index=False)

# Return path to new CSV
updated_file_path
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder # Import OneHotEncoder

# Load the dataset
df = pd.read_csv("theatre_seat_prediction_engineered.csv")

# Define features and target
target_seats = "total_seats"
features_seats = [col for col in df.columns if col not in [target_seats, "ticket_sold", "movie_name", "theatre_name"]]

# Train-test split
X_seats = df[features_seats]
y_seats = df[target_seats]
X_train, X_test, y_train, y_test = train_test_split(X_seats, y_seats, test_size=0.2, random_state=42)

# --- Categorical Feature Encoding ---
# Create a OneHotEncoder instance
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse=False for array output

# Fit the encoder on the categorical columns in the training data
categorical_cols = X_train.select_dtypes(include=['object']).columns
encoder.fit(X_train[categorical_cols])

# Transform the categorical features in both training and testing data
X_train_encoded = pd.DataFrame(encoder.transform(X_train[categorical_cols]), columns=encoder.get_feature_names_out(categorical_cols), index=X_train.index) # Added index=X_train.index
X_test_encoded = pd.DataFrame(encoder.transform(X_test[categorical_cols]), columns=encoder.get_feature_names_out(categorical_cols), index=X_test.index) # Added index=X_test.index

# Concatenate the encoded features with the numerical features
X_train = pd.concat([X_train.drop(categorical_cols, axis=1), X_train_encoded], axis=1)
X_test = pd.concat([X_test.drop(categorical_cols, axis=1), X_test_encoded], axis=1)
# --- End of Categorical Feature Encoding ---

# Train a model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Extract feature importance
feature_importance_df = pd.DataFrame({"Feature": X_train.columns, "Importance": model.feature_importances_}) # Use X_train.columns
feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)

# Plot Feature Importance
plt.figure(figsize=(12, 6))
sns.barplot(x=feature_importance_df["Importance"], y=feature_importance_df["Feature"], palette="viridis")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.title("Feature Importance for Predicting Theatre Capacity")
plt.show()

# Display top 10 features
print(feature_importance_df.head())
```
