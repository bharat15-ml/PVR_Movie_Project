```python
def dataset_details(dataset):
    feature = []
    dtype = []
    unique =[]
    count = []
    missing_values=[]
    missing_percentage = []

    for column in df.columns :
        feature.append(column)
        dtype.append(df[column].dtype)
        unique.append(df[column].unique())
        count.append(len(df[column]))
        missing_values.append(df[column].isnull().sum())
        missing_percentage.append(round((df[column].isnull().sum()/len(df))*100 , 2))


    details = pd.DataFrame({
        'Feature' : feature ,
        'Type' : dtype ,
        'Count' : count ,
        'Unique' : unique ,
        "Missed Values" : missing_values,
        'Missed Percent%' : missing_percentage,

    })

    return details
```

# Dealing with null values
    The idea is to replace null values of a particular city with the average ticket price of the non null values present in the city.
    For example for city Mumbai:
      Null Values(Mumbai) = Sum(Non-Null Values(Mumbai))/Length(Non-Null Values(Mumbai))

```python
from collections import defaultdict

total_price = defaultdict(lambda: 0)
length = defaultdict(lambda: 0)
#traversing through all the cities and calculating the sum and the length of non null values
for i in range(df.shape[0]):
    if pd.isna(df['average_ticket_price'][i]) == False:
        total_price[df['city'][i]] += df['average_ticket_price'][i]
        length[df['city'][i]] += 1

#replacing null values with resultant answer from above formula
for i in range(df.shape[0]):
    if pd.isna(df['average_ticket_price'][i]):
        avg = total_price[df['city'][i]]/length[df['city'][i]]
        df.loc[i, 'average_ticket_price'] = avg
```

# Top Ten Theatres in each City having Highest Average Ticket Price

```python
container = defaultdict(list)
#locations dataframe for next section
locations = pd.DataFrame(columns=['theatre_name', 'average_ticket_price', 'lat', 'lon'])

for i in range(df.shape[0]):
    dict1 = {}
    city = df['city'][i]
    theatre_name = df['theatre_name'][i]
    average_ticket_price = df['average_ticket_price'][i]
    lat = df['lat'][i]
    lon = df['lon'][i]
    dict1['theatre_name'] = theatre_name
    dict1['average_ticket_price'] = average_ticket_price
    dict1['lat'] = lat
    dict1['lon'] = lon

    container[city].append(dict1)
```

```python
cities = list(df['city'].unique())

for i in cities:
    newlist = sorted(container[i], key=lambda x: x['average_ticket_price'], reverse = True)
    theatre_name = []
    average_ticket_price = []
    for j in range(0,10):
        dict1 = newlist[j]
        theatre_name.append(dict1['theatre_name'])
        average_ticket_price.append(dict1['average_ticket_price'])
        locations.loc[len(locations)] = (dict1)
    plt.figure(figsize=(10,5))
    sns.set(style='whitegrid', color_codes = True)
    sns.barplot(x=theatre_name, y=average_ticket_price)
    plt.suptitle(i)
    plt.xticks(rotation=90)
    plt.show()
```
